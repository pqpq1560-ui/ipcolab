```python
from google.colab.patches import cv2_imshow
from google.colab import drive
import numpy as np
import cv2
import matplotlib.pyplot as plt
drive.mount('/content/drive')
```

```python
img = cv2.imread('/content/drive/My Drive/ip/Images/fce.bmp')
print(img.shape)
gImg = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
#print(gImg.shape)
cv2_imshow(gImg)
```

```python
kernel = np.ones((3,3),np.uint8)
erosion = cv2.erode(gImg,kernel,iterations = 1)
cv2_imshow(erosion)
```

```python
opening = cv2.dilate(erosion,kernel,iterations = 1)
cv2_imshow(opening)
```

```python
dilate2 = cv2.dilate(opening,kernel,iterations = 1)
cv2_imshow(dilate2)
```

```python
erode2 = cv2.erode(dilate2,kernel,iterations = 1)
cv2_imshow(erode2)
```
