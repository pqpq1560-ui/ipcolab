```python
img = cv2.imread('/content/drive/My Drive/ip/Images/milkdrop.bmp')
print(img.shape)
gImg = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
#print(gImg.shape)
cv2_imshow(gImg)
```

```python
kernel = np.ones((3,3),np.uint8)
dilatelmg = cv2.dilate(gImg,kernel,iterations = 3)
cv2_imshow(dilatelmg)
```

```python
gBin = np.zeros((gImg.shape[0],gImg.shape[1]), dtype= np.uint8)
```

```python
gBin = np.zeros((gImg.shape[0],gImg.shape[1]), dtype= np.uint8)
threshold = 150

for i in range(gImg.shape[0]):
  for j in range(gImg.shape [1]):
    if gImg[i,j] > threshold:
      gBin[i,j] = 255


cv2_imshow(gBin)
```

```python
kernel = np.ones((3,3),np.uint8)
dilatelmg = cv2.dilate(gBin,kernel,iterations = 3)
cv2_imshow(dilatelmg)
```

```python
kernel = cv2.getStructuringElement(cv2.MORPH_CROSS,(3,3))
dilatelmg = cv2.dilate(gBin, kernel,iterations = 3)
cv2_imshow(dilatelmg)
```
