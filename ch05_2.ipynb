```python
img = cv2.imread('/content/drive/My Drive/ip/Images/gsy2.jpg')
print(img.shape)
gImg = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
#print(gImg.shape)
cv2_imshow(gImg)
```

```python
outImg = cv2.blur(gImg,(7,7))
cv2_imshow(outImg)
```

```python
outImg = cv2.GaussianBlur(gImg,(7,7),5)
cv2_imshow(outImg)
```

```python
outImg = cv2.medianBlur(gImg,7)
cv2_imshow(outImg)
```

```python
outImage = gImg.copy()
#blurlmage = cv2.GaussianBlur(outImage,(7,7),5)
blurlmage = cv2.medianBlur(outImage, 7)

x = 200
y = 200
size = 15

for h in range(y, y+size):
  for w in range(x, x+size):
    #outImage[h][w] = 255
    outImage[h][w] =blurlmage[h][w]

cv2_imshow(outImage)
```

```python
img = cv2.imread('/content/drive/My Drive/ip/Images/lena.jpg')
print(img.shape)
gImg = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
#print(gImg.shape)
cv2_imshow(gImg)
```
